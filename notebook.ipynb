```python
df = collect_lightning_logs()

to_hide = ['final_epoch', 'best_valid_acc']
groups = df.groupby(["dataset", "strategy"])
# print("Groups:", list(groups.groups.keys()))

for name, group in groups:
    # drop unwanted columns before displaying (do not modify original df)
    group_to_show = group.drop(columns=[c for c in to_hide if c in group.columns])
    print(f"\n== {name} ({len(group)} rows) ==")
    display(group_to_show)
```
